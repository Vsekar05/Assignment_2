The Graph depicts the value of calories in the cereal dataset

The alpha value should be less than 0.05 which appears to be true

Based on the z test conducted on caloreis and carbo values in the cereal dataset bothe values appear to be independent.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as st
from statsmodels.stats.weightstats import ztest as ztest
tab = pd.crosstab(df['protein'],df['fat'])
tab = tab.T
print(tab)
k=st.chi2_contingency(tab)
print(k)
sns.countplot(df['fat'])
plt.show()
```

The Graph depicts the value of fat in the cereal dataset

```python
sns.countplot(df['protein'])
plt.show()
```

The Graph depicts the value of protein in the cereal dataset

The chi2 contigency test conducted on the values caloreis and carbo form the cereal dataset appears to be true.

The alpha value should be less than 0.05 which appears to be false

Based on the chi2 contigency test conducted on protein and fat values in the cereal dataset both values appear to be dependent.
